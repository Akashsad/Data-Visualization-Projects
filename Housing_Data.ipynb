```python
plt.figure(figsize= (10,10))
ax = sns.countplot(x = 'MSSubClass', data = dataset)
ax.bar_label(ax.containers[0])
plt.title("count of MSSubClass Houses")
plt.show()
```

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(x = 'MSZoning', data=dataset)
ax.bar_label(ax.containers[0])
plt.title("count of MSZoning Houses")
plt.show()
```

```python
dataset['LotShape'].value_counts().plot(kind='pie',autopct='%0.f%%')
plt.show()
```

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(x = 'LandContour', data=dataset)
ax.bar_label(ax.containers[0])
plt.title("Count of Flatness of the property")
plt.show()
```

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(x = 'LotConfig', data=dataset)
ax.bar_label(ax.containers[0])
plt.title("Count of Lot configuration")
plt.show()
```

```python
plt.figure(figsize = (5,5))
dataset['LandSlope'].value_counts().plot(kind='pie',autopct='%0.f%%')
plt.title('Count of Slope of property')
plt.show()
```

```python
plt.figure(figsize = (10,10))
dataset['Neighborhood'].value_counts().plot(kind='pie', autopct='%0.f%%')
plt.title("Count of location within ames city limits")
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.histplot(dataset['SalePrice'],kde=True, bins=30)
plt.title('Distribution of Sale Prices')
plt.xlabel('Sale price')
plt.ylabel('Number of Houses')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(dataset.corr(numeric_only=True),cmap='Paired',fmt='.2f',linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```
